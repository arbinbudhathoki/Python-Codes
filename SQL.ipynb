TRUNCATE vs DELETE
Truncate removes all rows instantly(where no clause allowed)
- IT deletes all rows at once, resets table to empty state and it is faster than delete.

TRUNCATE TABLE emp; # all data will remove instantly but the table structure will not 

This truncate will freeup the space which is occupied by the table.
No rollback is possible.




INSERT INTO emp Values(1, "NICK",2000)

AUTO increment resets is possible in Truncate
EXAMPLE: Clearing the entire classroom instantly, no selection everything is removed.
