```python

#Create Dataframe and add REVENUE column to it
import pandas as pd
data = {
    "Product":["Football","SoccerBall","Basketball"],
    "Price": ["10","16","18"],
    "Sold": ["5","8","19"]
}

df = pd.DataFrame(data, index = ["product1","product2"])
print(df)
```

```python
#add a new column revenue

df["revenue"] = df["Price"] * df["Sold"]
print(df)
```

```python
# add a record to the data
# add new record using LOC - to locate row
#to retirn one or more specified rows

df.loc[len(df)] = ["CricketBall",13,7, 1000*13] 
print(df)
```

```python
# Calculate total REvenue


total_revenue = df["revenue"].sum()
print(total_revenue)
```

```python
# loc[len(df)] - adds data at the next index position
df.loc[len(df)] = ["RemoteCar",100,1,100*1]
print(df)
```

```python
print(df.loc[[0,1]])
```

```python

#Introduce the Missing Vlaues
df.loc[2,"Sold"]= None
print(df)



```

```python
# Fill the missinf values
df["Sold"].fillna(0,inplace = True)
print(df)
```

```python
high_value = df[df["Price"]>15]
print(high_value)
```

```python
sorted_df = df.sort_values(by = "revenue", ascending = False )
print(sorted_df)
```

```python
import pandas as pd

data = {
    "Feeling": ["Happy", "Sad", "Broke"],
    "Price": [1200, 800, 500],
    "Sold": [12, None, 15]  
}


df = pd.DataFrame(data)
print(df)
```

```python
# detect the missing values - isnull
print(df.isnull())

#insull return True - value is missing, False if value is present
```

```python
print(df.isnull().sum())
```

```python
df.fillna(0, inplace = True)
print(df)

#Fillna(0) - it replaces NaN value with 0 or any alue you are passing, inplace  = True - It simple updates the original dataframe
```

```python
import pandas as pd

data = {
    "Feeling": ["Happy", "Sad", "Broke"],
    "Price": [1200, 800, 500],
    "Sold": [12, None, 15]  
}


df = pd.DataFrame(data)

# Fill missing values with mean and average
df["Sold"].fillna(df["Sold"].mean(),inplace = True)
print(df)
```
