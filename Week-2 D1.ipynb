```python
# Add 10 to each number
# using For loop
numbers = [1,2,3,4,5]
result = []
for num in numbers:
    result.append(num + 10)

print(result)
```

```python
number = [2,3,4,5,6]    #MAP function
def add_ten(x):
    return x + 10

result = map(add_ten, number)   #applies function to each element
print(list(result))
```

```python
#you have the salary data -> need to add 10% bonus using map function
sal = [10000,20000,30000]
def add_sal(x):
    return x + (x*0.10)
result = map(add_sal, sal)
print(list(result))
```

```python
# base case - condition to stop the recursion, recursive case - when function calling itself.

#Print numbers from 1 to 5

def print_numbers(n):
    if n == 6: # base case
        return
    print(n)
    print_numbers(n+1)   #This is recursive call

print_numbers(1)







```

# Fibonacci

```python
# Fibonacci Series
#Sequence where you have 0,1,1,2,3,5,8,13
f(n)= f(n-1) + f(n-2)
```

```python
def fib(n):
    if n ==0:
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1) + fib(n-2)

print (fib(5))

```
