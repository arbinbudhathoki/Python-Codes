```python
#Inheritance
class Person:    
    def __init__(self,name): 
        self.name = name
    def show_name(self):
        print("Name: ", self.name)

class Employee(Person):  #intertaning Person
    def __init__(self,name,salary):
        super().__init__(name)    # Calling parent constructor
        self.salary = salary 
        

# Super callMethod - call Methods of the Parent clas, to initialize parent properties, to reuse parent logic      
# It allows you to call methods from a parent class without explicitly naming it.

emp = Employee("Arbin",40000)
  #inherited method
print(emp.salary)
print(emp.name)
```

```python
 # Polymorphism
# 1 - Method Overriding - when child class provides its own implementation of a mathod already defined in  a parent class
# 
#
class Parent:
    def show(self):
        print("This is a crazy person.")

class Employee(Parent):
    def show(self):    # Overriding parent Method
        super().show()  #call parent methof - Parent Logic + child Logic
        print("This is an Employee.")

obj = Employee()
obj.show()

#Python checks: First -> child class Then - > child method overrides parent method





     
```

```python
# Method Overloading - same function name with different number/ type of arguement - default arguement *args

def add(a,b=0,c=0):
    return a+b+c


#same function bheaves differently based on inputs
print(add(10))
print(add(10,20))
print(add(10,20,30))








```
