# While LOOP


```python
number = 0
while number <5:
    print(number)
    number = number +1
```

```python
number = 0
while number <5:
    
    number = number +1
    print(number)
```

```python
number = 0
while number <5:
    print(number)
    if number ==4:
        break
    number = number +1
    
```
